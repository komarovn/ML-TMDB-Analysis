```python
data_mod = data.drop(['id', 'original_title', 'vote_count', 'popularity', 'homepage', 'keywords', 'genres', 'overview', 'production_companies', 'production_countries', 'release_date', 'spoken_languages', 'status', 'tagline', 'title'], axis = 1)
```

```python
categorical_columns = [c for c in data_mod.columns if data_mod[c].dtype.name == 'object']
numerical_columns   = [c for c in data_mod.columns if data_mod[c].dtype.name != 'object']
print (categorical_columns)
print (numerical_columns)
```

```python
data_numerical = data_mod[numerical_columns]
data_numerical.describe()
```

Нормализуем

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
data_numerical.describe()
```

```python
for c in categorical_columns:
    print(c, data_mod[c].unique())
```

```python
data_nonbinary = data_mod['original_language']
```

# X и y

```python
X = data_res.drop(('revenue'), axis = 1) # выбрасываем столбец 'revenue'
y = data_res['revenue']
feature_names = X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

# kNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print ('Error train:', err_train)
print ('Error test:', err_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

```python
confusion_matrix([0,1,0,1,0], [0,1,0,0,0])
```

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

# ДЗ 2

Обучение классификатора

```python
from sklearn.neural_network import MLPClassifier
```

Возьмём 5 нейронов

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (5), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
```

Ошибка на обучающей и тестовой выборках

```python
print(np.mean(y_train != y_train_pred),
    np.mean(y_test != y_test_pred))
```

Ошибка на тестовой выборке больше, чем на обучающей

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (3), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred),
    np.mean(y_test != y_test_pred))
```

Добавим слой

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (3,4), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred),
    np.mean(y_test != y_test_pred))
```

Двуслойная нейронная сеть показывает лучшие результаты

Подбор гиперпараметров

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (3,4), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Найдём оптимальное значение для alpha

```python
alpha_opt = alpha_arr[test_err == min_test_err]
alpha_opt = alpha_opt[0]
print(alpha_opt)
```

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (3,4), solver = 'lbfgs',
                          activation = 'logistic', random_state = 5)
mlp_model.fit(X_train, y_train)
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred),
     np.mean(y_test != y_test_pred))
```

Ошбика уменьшилась

Найдём оптимальное количество нейронов

```python
neuron_arr = np.arange(1,30)
test_err = []
train_err = []
for neuron in neuron_arr:
    mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (neuron,neuron), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(neuron_arr, train_err, 'b-o', label = 'train')
plt.semilogx(neuron_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.min(neuron_arr), np.max(neuron_arr)])
plt.title('Error vs. neurons')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Минимальная ошибка при 4-х нейронах.

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (4,4),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred),
      np.mean(y_test != y_test_pred))
```

Матрица рассогласования и другие метрики качества

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_pred, y_test))
```

```python
M = confusion_matrix(y_test_pred, y_test)
M = np.sqrt(M)
plt.imshow(M, interpolation = 'nearest')
plt.set_cmap('binary')
plt.xticks(range(2))
plt.yticks(range(2))
plt.xlabel("true label")
plt.ylabel("predicted label")
plt.colorbar()
pass
```

```python
counter = 0
for i in X_test.index:
    x = X_test.loc[i, :]
    dig = y_test[i]
    dig_pred = mlp_model.predict([x])
    if dig_pred != dig:
        counter += 1
print("Количество ошибок: "+ str(counter))
print("Размер тестовой выборки: " + str(len(X_test)))
```

# ДЗ 3

### Random Forest

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)
```

```python
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
print(np.mean(y_train != y_train_pred), np.mean(y_test != y_test_pred))
```

Ошибка на тестовой больше, чем на обучающей. Имеем дело с переобучением.

```python
trees_arr = np.arange(30, 200, 10)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = trees, oob_score=True)
    rf.fit(X_train, y_train)
    train_err.append(np.mean(y_train != rf.predict(X_train)))
    test_err.append(np.mean(y_test  != rf.predict(X_test)))
    oob_err.append(1 - rf.oob_score_)

plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. Trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print( min_train_err)
print( min_test_err)
```

Оптимальное количество деревьев

```python
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt)
```

### Extremely Random Forest

```python
ert = ensemble.ExtraTreesClassifier(n_estimators = 100)
ert.fit(X_train, y_train)
```

```python
y_train_pred = ert.predict(X_train)
y_test_pred = ert.predict(X_test)
print(np.mean(y_train != y_train_pred), np.mean(y_test != y_test_pred))
```

Наблюдается переобучение

```python
trees_arr = np.arange(50, 200, 10)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    ert = ensemble.ExtraTreesClassifier(n_estimators = trees, oob_score=True, bootstrap=True)
    ert.fit(X_train, y_train)
    train_err.append(np.mean(y_train != ert.predict(X_train)))
    test_err.append(np.mean(y_test  != ert.predict(X_test)))
    oob_err.append(1 - ert.oob_score_)
    
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. Trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Оптимальное количество деревьев

```python
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt)
```
